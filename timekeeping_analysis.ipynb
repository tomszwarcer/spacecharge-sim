```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("timekeeping.txt",header=None)
df.columns = ["on","steps","time"]
on = df[df["on"]==1]
off = df[df["on"]==0]
steps = [40,80,160]

x = np.arange(3)
y1 = [on[on["steps"]==i]["time"].mean() for i in steps]
y2 = [off[off["steps"]==i]["time"].mean() for i in steps]
s1 = [on[on["steps"]==i]["time"].std() for i in steps]
s2 = [off[off["steps"]==i]["time"].std() for i in steps]

width = 0.4
plt.bar(x-width/2,y1,width=width,label = "Optimised")
plt.bar(x+width/2,y2,width=width,label = "Original")
plt.xticks(x, ['40', '80', '160'])
plt.xlabel("Number of grid steps")
plt.ylabel("Mean CPU time (s)")
plt.title("Comparison of optimisation")
plt.legend()
plt.savefig("figures/allbars.png")
```
