```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

# No SC

```python
n_bins = 10
infile = open("size_no_sc_combined.txt")
l  = infile.readlines()
l = [i[:-1].split(",") for i in l]
voltages = [480,500,520,540,560]
gain_list = []
print(l)

for i in voltages:
    gain_list_voltage = []
    for entry in l:
        if entry[0] == str(i):
            gain_list_voltage.append(int(entry[1]))
    gain_list.append(gain_list_voltage)
print(gain_list)
```

```python
mean_gains = []
errors = []

for i in range(len(gain_list)):  
  #calculate mean
  total_gain = 0
  num_runs = len(gain_list[i])
  for gain in gain_list[i]:
    total_gain += gain
  mean = total_gain/num_runs

  #calculate variance
  var = 0
  for gain in gain_list[i]:
    var += (gain - mean)**2
  var = var/num_runs
  se = np.sqrt(var)/np.sqrt(num_runs)

  mean_gains.append(mean)
  errors.append(se)


for i in range(len(voltages)):
  #Plot histogram
  entries, bc, p = plt.hist(gain_list[i], bins=n_bins, density=False)
  plt.xlabel("Gain")
  plt.ylabel("Counts")
  plt.title("Gain distribution [dV = "+str(voltages[i])+"]")
  plt.text(x=0.5*plt.xlim()[1],y=0.75*plt.ylim()[1],s="mean = " + str(int(round(mean,0))) + "\nSD = " + str(int(round(se,0))))
  plt.savefig("plots/no_sc_"+str(voltages[i])+".png")
  plt.close()
  
```
